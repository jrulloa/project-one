```python
import pandas as pd
```

###Population

```python
population_filepath = "Resources/Population.csv"
income_filepath = "Resources/personal_income (USD).csv"
gdp_filepath = "Resources/GDP_USD.csv"
expenditure_filepath = "Resources/spending_per_category.csv"
```

```python
population_df = pd.read_csv(population_filepath)
# population_df.head()
population_clean = population_df[['GeoName', '2014', '2015', '2016', '2017', '2018', '2019']]
# population_clean
```

```python
#Cleaning the Population Clean
population_cleaner = population_clean.dropna(how='any')
# population_cleaner
```

```python
states_df = population_cleaner.iloc[1:51]
# states_df
```

```python
final_pop_df = states_df.rename(columns={"GeoName": "States"})
# final_pop_df
```

###Income Data

```python
income_df = pd.read_csv(income_filepath)
# income_df.head()
```

```python
income_clean = income_df[['GeoName', '2014', '2015', '2016', '2017', '2018', '2019']]
# income_clean
income_cleaner = income_clean.dropna(how='any')
# income_cleaner
states_income_df = income_cleaner.iloc[1:51]
# states_income_df
final_inc_df = states_income_df.rename(columns={"GeoName": "States"})
# final_inc_df
```

###GDP

```python
gdp_df = pd.read_csv(gdp_filepath)
# gdp_df.head()
```

```python
gdp_clean =gdp_df[['GeoName', '2014', '2015', '2016', '2017', '2018', '2019']]
# income_clean
gdp_cleaner = gdp_clean.dropna(how='any')
# income_cleaner
states_gdp_df = gdp_cleaner.iloc[1:51]
# states_income_df
final_gdp_df = states_gdp_df.rename(columns={"GeoName": "States"})
# final_gdp_df
```

```python
final_pop_df["States_Pop_Average"]=final_pop_df.mean(axis=1)
final_pop_df

```

```python
final_inc_df["States_Inc_Average"]=final_inc_df.mean(axis=1)
final_inc_df
```

```python
final_gdp_df["States_GDP_Average"]=final_gdp_df.mean(axis=1)
final_gdp_df
```

###Merging datasets

```python
merge_pop_inc_df = pd.merge(final_pop_df, final_inc_df, how="left", on="States")
merge_pop_inc_df
```

```python
merge_pop_inc_gdp_df = pd.merge(merge_pop_inc_df, final_gdp_df, how="left", on="States")
merge_pop_inc_gdp_df 
```

```python
merge_pop_inc_gdp_df.columns
```

```python
merge_pop_inc_gdp_df = merge_pop_inc_gdp_df.rename(columns={'2014_x': '2014_Pop', '2015_x':'2015_Pop', '2016_x':
                                                            '2016_Pop', '2017_x':'2017_Pop', '2018_x':'2018_Pop',
                                                            '2019_x':'2019_Pop',
                                                            '2014_y': '2014_Inc', '2015_y': '2015_Inc', 
                                                            '2016_y': '2016_Inc', '2017_y': '2017_Inc', '2018_y': '2018_Inc',
                                                            '2019_y': '2019_Inc', 
                                                            '2014': '2014_GDP', '2015': '2015_GDP',
                                                            '2016': '2016_GDP', '2017': '2017_GDP', '2018': '2018_GDP', '2019': '2019_GDP'})
merge_pop_inc_gdp_df
```

###Calculate average for POP from 2014-2019 per state, for INC from 2014-2019 per state, for GDP from 2014-2019 per state

###Use average per state to calculate the Income and GDP per capita

```python
per_capita = merge_pop_inc_gdp_df[["States","States_Pop_Average", "States_Inc_Average", "States_GDP_Average"]]
per_capita


per_capita['per_capita_inc'] = (per_capita['States_Inc_Average']/per_capita['States_Pop_Average'])*1000000
per_capita['per_capita_gdp'] = (per_capita['States_GDP_Average']/per_capita['States_Pop_Average'])*1000000
per_capita

```

###Grab top 5 and lowest 5 states for per capita GDP and per capita income

```python
sorted_inc = per_capita.sort_values("per_capita_inc", ascending=True)
sorted_inc= sorted_inc[["States", "per_capita_inc"]]

sorted_inc= sorted_inc.reset_index(drop=True)
sorted_inc
```

```python
bottom_inc = sorted_inc.iloc[0:5]
bottom_inc

```

```python
top_inc = sorted_inc.iloc[45:]
top_inc
```

```python
sorted_gdp = per_capita.sort_values("per_capita_gdp", ascending=True)
sorted_gdp= sorted_gdp[["States", "per_capita_gdp"]]

sorted_gdp= sorted_gdp.reset_index(drop=True)
sorted_gdp
```

```python
bottom_gdp = sorted_gdp.iloc[0:5]
bottom_gdp

```

```python
top_gdp = sorted_gdp.iloc[45:]
top_gdp
```
