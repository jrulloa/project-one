```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
```

###Population

```python
population_filepath = "Resources/Population.csv"
income_filepath = "Resources/personal_income (USD).csv"
gdp_filepath = "Resources/GDP_USD.csv"
expenditure_filepath = "Resources/spending_per_category_by_year.csv"
gas_filepath = "Resources/gas.csv"
```

```python
population_df = pd.read_csv(population_filepath)
# population_df.head()
population_clean = population_df[['GeoName', '2015', '2016', '2017', '2018', '2019']]
# population_clean
```

```python
#Cleaning the Population Clean
population_cleaner = population_clean.dropna(how='any')
# population_cleaner
```

```python
states_df = population_cleaner.iloc[1:52]
# states_df
```

```python
final_pop_df = states_df.rename(columns={"GeoName": "States"})
# final_pop_df
```

###Income Data

```python
income_df = pd.read_csv(income_filepath)
# income_df.head()
```

```python
income_clean = income_df[['GeoName', '2015', '2016', '2017', '2018', '2019']]
# income_clean
income_cleaner = income_clean.dropna(how='any')
# income_cleaner
states_income_df = income_cleaner.iloc[1:52]
# states_income_df
final_inc_df = states_income_df.rename(columns={"GeoName": "States"})
# final_inc_df
```

###GDP

```python
gdp_df = pd.read_csv(gdp_filepath)
# gdp_df.head()
```

```python
gdp_clean =gdp_df[['GeoName', '2015', '2016', '2017', '2018', '2019']]
# income_clean
gdp_cleaner = gdp_clean.dropna(how='any')
# income_cleaner
states_gdp_df = gdp_cleaner.iloc[1:52]
# states_income_df
final_gdp_df = states_gdp_df.rename(columns={"GeoName": "States"})
# final_gdp_df
```

```python
final_pop_df["States_Pop_Average"]=final_pop_df.mean(axis=1)
final_pop_df

```

```python
final_inc_df["States_Inc_Average"]=final_inc_df.mean(axis=1)
final_inc_df
```

```python
final_gdp_df["States_GDP_Average"]=final_gdp_df.mean(axis=1)
final_gdp_df
```

###Merging datasets

```python
merge_pop_inc_df = pd.merge(final_pop_df, final_inc_df, how="left", on="States")
merge_pop_inc_df
```

```python
merge_pop_inc_gdp_df = pd.merge(merge_pop_inc_df, final_gdp_df, how="left", on="States")
merge_pop_inc_gdp_df 
```

```python
merge_pop_inc_gdp_df.columns
```

```python
merge_pop_inc_gdp_df = merge_pop_inc_gdp_df.rename(columns={'2015_x':'2015_Pop', '2016_x':
                                                            '2016_Pop', '2017_x':'2017_Pop', '2018_x':'2018_Pop',
                                                            '2019_x':'2019_Pop',
                                                            '2015_y': '2015_Inc', 
                                                            '2016_y': '2016_Inc', '2017_y': '2017_Inc', '2018_y': '2018_Inc',
                                                            '2019_y': '2019_Inc', 
                                                            '2015': '2015_GDP',
                                                            '2016': '2016_GDP', '2017': '2017_GDP', '2018': '2018_GDP', '2019': '2019_GDP'})
merge_pop_inc_gdp_df
```

###Calculate average for POP from 2015-2019 per state, for INC from 2015-2019 per state, for GDP from 2015-2019 per state

###Use average per state to calculate the Income and GDP per capita

```python
per_capita = merge_pop_inc_gdp_df[["States","States_Pop_Average", "States_Inc_Average", "States_GDP_Average"]]
per_capita


per_capita['per_capita_inc'] = (per_capita['States_Inc_Average']/per_capita['States_Pop_Average'])*1000000
per_capita['per_capita_gdp'] = (per_capita['States_GDP_Average']/per_capita['States_Pop_Average'])*1000000
per_capita

```

```python
#Getting gas Cost
gas_cost_df = pd.read_csv(gas_filepath)
gas_cost_df = gas_cost_df[['Year','Price per gallon','State Name']]
gas_cost_df = gas_cost_df.rename(columns= {'State Name':'State'})
gas_cost_df = gas_cost_df.sort_values('Year', ascending=True)
gas_cost_df
```

```python
#Merge Expenditures and gas cost

gas_cost_Expend_df = pd.merge(expenditure_df,gas_cost_df, how="left", on= ["State" , "Year"] )
gas_cost_Expend_df

#PriceGas= gas_cost_Expend_df.iloc[:,2]
#PriceGas
```

###Plotting Data gas price vs. durable goods


```python
x = gas_cost_Expend_df['Price per gallon'] # Gasoline Price 
y_dg = gas_cost_Expend_df.iloc[:,2] #Durable Goods
y_ndg = gas_cost_Expend_df.iloc[:,3] #NO Durable Goods
y_s = gas_cost_Expend_df.iloc[:,4] # Services

fignum = 1

#Linear Regression Durable Goods
(slope, intercept, rvalue, pvalue, stderr) = linregress(x,y_dg)
r2 =(f"The r-squared is: {round(rvalue**2,4)}")
regress_values = x * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(f"The slope of the line is:{round(slope,3)}")
print(f"The y-intercept is:{round(intercept,3)}")
print(f"The rvalue (Correlation coefficient) is:{round(rvalue,3)}")
print(f"The pvalue is:{round(pvalue,3)}")
print(f"The standard error is:{round(stderr,3)}")

plt.scatter(x,y_dg, marker="o", facecolors="green", edgecolors="black",s=30, alpha=0.75)
plt.title("Durable Goods")
plt.plot(x,regress_values,"r-")
plt.annotate(line_eq,(2.7,140000),fontsize=12,color="red")
plt.annotate(r2,(2.7,125000),fontsize=10,color="black")

plt.savefig(f"output_data/fig{fignum}.png")
fignum = fignum + 1 

```
